## 1. Introduction(도입부)
### 1.1 Data Dictionary(분석할 변수(요인))

##### 1)<font color=#CD0000>Age</font> : Age of the patient <font color=#00008C>(환자 나이)</font><br>

##### 2)<font color=#CD0000>Sex</font> : Sex of the patient (1 = male, 0 = female) <font color=#00008C>(환자 성별)</font><br>

##### 3)<font color=#CD0000>exang</font> : exercise induced angina (1 = yes; 0 = no) <font color=#00008C>(운동으로 인한 협심증 여부(1 = 예, 0 = 아니오))</font><br>

##### 4)<font color=#CD0000>caa</font> : number of major vessels (0-3) <font color=#00008C>(형광체로 착색된 주요 혈관의 수(0-3개))</font><br>

##### 5) <font color=#CD0000>cp</font> : Chest Pain type chest pain type <font color=#00008C>(흉통 유형)</font>
  - Value 1: typical angina(전형적 협심증)<br>
  - Value 2: atypical angina(비정형 협심증)<br>
  - Value 3: non-anginal pain(비 협심증)<br>
  - Value 4: asymptomatic(무증상 협심증)<br>

##### 6) <font color=#CD0000>trtbps</font> : resting blood pressure (in mm Hg) <font color=#00008C>(휴식 혈압 (mm Hg))</font><br>

##### 7) <font color=#CD0000>chol</font> : cholestoral in mg/dl fetched via BMI sensor <font color=#00008C>(혈청 콜레스테롤 (mg / dl))</font><br>

##### 8) <font color=#CD0000>fbs</font> : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) <font color=#00008C>(공복 혈당 (120mg / dl))</font><br>

##### 9) <font color=#CD0000>rest_ecg</font> : resting electrocardiographic results <font color=#00008C>(휴식 심전도 결과)</font>
  - Value 0: normal(정상)<br>
  - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)(ST-T 파 이상)<br>
  - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria(좌심실 비대)<br>
  
#### 10) <font color=#CD0000>thalach</font> : maximum heart rate achieved<font color=#00008C>(탈륨 스트레스 테스트 중 달성 된 심박수 최대치)</font><br>

##### 11) <font color=#CD0000>output</font> : 0= less chance of heart attack <font color=#00008C>(심장마비에 걸릴 낮은 확률)</font> 1= more chance of heart attack <font color=#00008C>(심장마비에 걸릴 높은 확률)</font><br>

### 4.2 t-test (그룹 간 평균의 비교)

```python
#output과 cp, caa, thalachh, sex, 그리고 exang에서 유의미한 결과를 도출해낼 수 있었으므로 다음 3가지 요인과 결과값을 검정해보기로 함.
#남성과 여성의 나이를 비교해보자
df.groupby('sex').output.mean()
```

```python
# 그룹간 평균 비교: 2 sample t-test (independent t-test) 
#1. 검정통계량(많이 벗어날 수록 통계에 유리), 두데이터가 동일하다고 생각했을 때 나올 수 있는 값(귀무가설(두 데이터가 차이가 없을 때 나타나는 값)과 대입가설(두 데이터가 차이가 있다고 가정할 때 나올 가능성)에서 확인할 수 있다.)
[t_stat, p] = stats.ttest_ind(df[df.sex == 1].output, df[df.sex == 0].output)
[t_stat, p]
#위 결과 값을 보면, 현재 데이터만으로 남성과 여성이 유의미한 차이를 가진 데이터라고 말하기 어렵다.
```

### 4.4 카이 제곱 검정(교차표(cross-tab) 생성 및 그룹간 빈도 비교)

```python
# pd.cut() 구간화 기능을 이용한 연령 분류
age_cut = [20, 40, 60, 80]    # [20~40): 청년, [40~50): 중년, [60~80대): 노년
df['age_range'] = pd.cut(df.age, age_cut, labels = ['청년', '중년', '노년'], right = False)
df
```

```python
# 교차표 생성: pd.crosstab()
pd.crosstab(df.sex, df.age_range)
```

```python
# 그룹간 빈도 비교: 카이 제곱 검정 (chi-square test) 
crosstab = pd.crosstab(df.sex, df.age_range)
stats.chi2_contingency(crosstab)#이런 값이 우연히 나올 가능성은 낮다-> 남성과 여성과의 체계적인 차이가 존재할 것이다. 
```

6. EDA
이제 본격적으로 데이터를 다뤄 보겠습니다.


제가 다루려는 데이터는 변수가 무척이나 많았으므로 우선 저희가 배운 다변량 분석을 이용해서 그래프를 그리는 등 분석을 진행했습니다.

다음 그림은 변수별로 나타날 수 있는 수치를 그래프로 정리한 것입니다.

여기 성별 부분에서 0은 여성을, 1을 남성을 의미하고 있습니다. 
이 그래프를 본다면 지금 다루는 데이터는 여성보다 남성의 것이 더 많다는 것을 알 수 있습니다.


그 다음 그림은 각 변수값의 분포를 boxen plot이라는 것으로 나타낸 것인데, 저희가 전에 배웠던 box plot과 violin plot을 섞어 놓은 듯한 형태라 흥미로울 것 같아 이렇게 나타내 보았습니다. 

여기 제일 우측 상단에 있는 trtbps는 '휴식 혈압'을 나타낸 것인데 그래프 위쪽의 꼬리 부분이 많이 올라간 것을 보아 이상치가 꽤 많이 존재한다는 것을 알 수 있고, 마찬가지로 다른 변수들도 이상치를 많이 가지고 있다는 것을 알 수 있었습니다.

다음 그래프는 환자들이 심장마비에 걸릴 가능성이 높으면 1, 낮다면 0을 부여한 output column을 그래프로 나타낸 것입니다.

심장마비에 걸릴 가능성이 높은 환자가 더 많긴 하지만 그렇게 가능성이 낮은 환자와 수가 그리 차이나지는 않는 다는 것을 알 수 있고,
계속 데이터 분석 진행해보도록 하겠습니다.

다음은 2가지 변수끼리의 상관관계를 알아보는 이변량 분석을 해보도록 하겠습니다.

corr()함수를 통해 상관관계 매트릭스를 만들어 변수끼리 유의미한 상관관계를 가지고 있는지 관찰합니다. 

아래 그림을 참고해 상관관계를 판단하면,
age와 thalachh, chol과 oldpeak, thalachh와 oldpeak이 각각 뚜렷한 양적 상관관계를 가지고 있다는 결론을 도출할 수 있었습니다.

다음 매트릭스는 한번도 본 적이 없는 형태의 도식이어서 가져와봤는데요, 위의 상관관계를 매트릭스로 나타낸 것입니다. 
두 변수가 양의 상관관계를 가질수록 파란색에 가까워지고, 음의 상관관계를 가질 수록 연노란색에 가까워지는 것을 확인할 수 있습니다.  

그 다음 매트릭스도 같은 상관관계를 다른 방법으로 도식화한 것인데요, 상관관계가 크수록 원의 크기가 커지고 파란색에 가까워집니다.


이제 output과 각 변수 간의 관계를 알 수 있는 그래프입니다.
age의 경우, 심장 마비에 걸릴 가능성이 높은 사람들의 수치가 낮은 사람들의 수치보다 높은 값을 나타내는 것을 알 수 있습니다.
이 그래프로 짐작해본다면 나이대가 높을수록 심장마비에 걸릴 가능성이 높다고 예측할 수 있습니다.

만들어진 그래프들 중 chol(혈청 콜레스테롤), thalachh(심박수 최대치), cp(흉통 유형)이 유의미한 결과값의 차이를 보이는 것을 관찰할 수 있습니다.

마지막으로 변수들 간의 관계를 pairplot으로 나타낸 것입니다.

5) 스캐터 플롯을 통해 output과 cp, 그리고 thalachh에서 상관관계를 확인할 수 있다.

6) 노인들이 심장마비가 올 가능성이 더 높다고 예상하지만, 연령대별 심장마비 분포도를 보면 그렇지 않다.

7) thalachh 분포도로 최대 심장 박동수가 높은 사람은 심장마비의 가능성이 더 높다는 것을 알 수 있다.

8) 3.2.4 Some other relations that seemed intuitive(예상했던 다른 요인들과의 관계)에서 도출할 수 있는 분석 결과

- 비혈관통증, 즉 cp = 2인 사람은 심장마비가 올 확률이 높다.

- 남성(sex = 1)이 심장마비 발생 확률이 더 높다.
- 운동으로 인한 협심증이 없는 사람, 즉 exang = 0일 때 심장마비가 올 확률이 높다.
