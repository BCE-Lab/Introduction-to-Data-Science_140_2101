## 국내총생산(Gross Domestic Product, GDP)은 한 나라의 영역 내에서 가계, 기업, 정부 등 모든 경제주체가 일정기간 동안 생산한 재화 및 서비스의 부가가치를 시장가격으로 평가하여 합산한 것으로 여기에는 비거주자가 제공한 노동, 자본 등 생산요소에 의하여 창출된 것도 포함되어 있다.

```python
consumer = pd.read_csv('가계의_목적별_최종소비지출_분기.csv',encoding = 'CP949')
consumer = consumer.set_index('가계의 목적별 최종소비지출')
consumer
```

```python
#2008 3/4분기 ~ 2009 3/4분기를 뽑아내서 분석
c_20082009 = consumer.iloc[:13,4*4-2:4*5-1]
c_20082009
C_1=[]
for i in range(4):
    C_1.append(c_20082009.iloc[:,i+1]/c_20082009.iloc[:,i])
    
m_c = pd.concat([pd.DataFrame(C_1[0]),pd.DataFrame(C_1[1]),pd.DataFrame(C_1[2]),pd.DataFrame(C_1[3])],axis=1)
m_c

def test(x):
        if x<1: 
            return float((1-x)*-100)
        elif x>2:
            return float(x*100)
        else: return float((x-1)*100)
        
m_c['비율(%)_20084/4']= m_c.iloc[:,0].apply(test)
m_c['비율(%)_20091/4']= m_c.iloc[:,1].apply(test)
m_c['비율(%)_20092/4']= m_c.iloc[:,2].apply(test)
m_c['비율(%)_20093/4']= m_c.iloc[:,3].apply(test)
m_c = m_c.iloc[:,4:]
m_c
```

```python
m_c.plot(kind='bar',figsize = (20,10))
```

이 그래프를 그릴 때 경기후퇴기 앞뒤로 하나씩 더 분기를 추가하여 그래프를 그렸다. 
경기후퇴기에 연속감쇠를 보인 분야는 식료품 및 비주류음료, 주류 및 담배, 의류 및 신발 분야이다. 이에 사람들이 2008_2009 경기 후퇴기에서는 이전에 비해 같은 분야에서의 소비가 줄었다는 것을 확인할 수 있다.

```python
#2019 3/4분기 ~ 20204/4분기 뽑아내서 분석
c_20192020 = consumer.iloc[:13,4*15-2:4*16+1]
c_20192020
C_2=[]
for i in range(5):
    C_2.append(c_20192020.iloc[:,i+1]/c_20192020.iloc[:,i])
C_2
m_c2 = pd.concat([pd.DataFrame(C_2[0]),pd.DataFrame(C_2[1]),pd.DataFrame(C_2[2]),pd.DataFrame(C_2[3]),pd.DataFrame(C_2[4])],axis=1)
def test(x):
        if x<1: 
            return float((1-x)*-100)
        elif x>2:
            return float(x*100)
        else: return float((x-1)*100)
        
m_c2['비율(%)_20194/4']= m_c2.iloc[:,0].apply(test)
m_c2['비율(%)_20201/4']= m_c2.iloc[:,1].apply(test)
m_c2['비율(%)_20202/4']= m_c2.iloc[:,2].apply(test)
m_c2['비율(%)_20203/4']= m_c2.iloc[:,3].apply(test)
m_c2['비율(%)_20204/4']= m_c2.iloc[:,4].apply(test)
m_c2 = m_c2.iloc[:,5:]
m_c2
```

```python
m_c2.plot(kind='bar',figsize = (20,10))
```

이 그래프를 그릴 때 경기침체기 앞에서 한 분기를 추가하여 그래프를 그렸다. 
코로나 유행으로 인해 소비 분야에서 뚜렷한 감소를 보인 부분은 교육서비스 분야이다. 
코로나로 인해 사람이 모이는 상황을 피하다보니 학원이나 과외를 다니는 것을 중단하거나 다니려고 했던 학원을 안다니는 등 교육 서비스에 대한 소극적인 소비가 나타났다는 것을 확인할 수 있었다.
