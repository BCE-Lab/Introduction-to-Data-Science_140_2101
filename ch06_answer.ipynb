```python
# 아래의 라이브러리를 사용하므로 미리 임포트해 두기 바랍니다.
import numpy as np
import numpy.random as random
import scipy as sp
import scipy.stats as stats
import pandas as pd
from pandas import Series, DataFrame

# 시각화 라이브러리
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

# 소수점 3자리까지 표시
%precision 3

data = pd.read_csv('health2018.csv')
data
```

## 1. 판다스를 이용한 데이터 조작
### 1-1. 구간화(pd.cut)를 이용해 다음 체중 분류를 obesity 열로 추가하시오.
BMI 18.5 미만 : "저체중"

BMI 18.5 이상 25 미만 : "정상"

BMI 25 이상 30 미만: "비만"

obesity 열을 추가한 후 data.head(30)으로 처음 30개 데이터를 체크하시오.

```python
BMI_cut = [0, 18.5, 25, 100]
BMI_names = ['저체중', '정상', '비만']
data['obesity'] = pd.cut(data.BMI, BMI_cut, labels = BMI_names)
data.head(30)
```

### 1-2. 구간화(pd.cut)를 이용해 다음 연령 분류를 age_group 열로 추가하시오.
20세 이상 30세 미만: '20대'

30세 이상 40세 미만: '30대'

40세 이상 50세 미만: '40대'

50세 이상 60세 미만: '50대'

60세 이상 70세 미만: '60대'

70세 이상 80세 미만: '70대'

```python
age_cut = [20, 30, 40, 50, 60, 70, 80]
age_names = ['20대', '30대', '40대', '50대', '60대', '70대']
data['age_group'] = pd.cut(data.age, age_cut, labels = age_names, right = False)
data
```

### 1-3. 매핑 기능을 활용해서 다음과 같이 age_group2 열을 생성하라.
20대, 30대 : 청년 

40대, 50대 : 중년 

60대, 70대 : 노년 

```python
age_map = {
    '20대': '청년',
    '30대': '청년',
    '40대': '중년',
    '50대': '중년',
    '60대': '노년',
    '70대': '노년'
}
data['age_group2'] = data['age_group'].map(age_map)
data
```

## 2. 통계적 가설 검정
### 2-1. 20대 그룹에서 정상 체중과 비만 체중인 사람의 혈당값 평균을 구하고, 통계적으로 유의한 차이(p < 0.05)가 나는지 검정하시오.

```python
data20 = data[(data.age_group == '20대')]
print(data20.groupby('obesity').Blood_glucose.mean())
[t_stat, p] = stats.ttest_ind(data20[data20.obesity == '정상'].Blood_glucose, data20[data20.obesity == '비만'].Blood_glucose)
[t_stat, p]
```

### 2-2. 20대 남성에서 정상 체중과 비만 체중인 사람간에 키 차이가 통계적으로 유의한지 검정해보시오.

```python
data20M = data[(data.age_group == '20대') & (data.sex == 'M')]
print(data20M.groupby('obesity').height.mean())
[t_stat, p] = stats.ttest_ind(data20M[data20M.obesity == '정상'].height, data20M[data20M.obesity == '비만'].height)
[t_stat, p]
```

### 2-3. 청년(20~30대) 그룹에서 성별로 저체중/정상/비만 체중 분포를 교차표로 구하시오.

```python
pd.crosstab(data[data.age_group2 == '청년'].sex, data[data.age_group2 == '청년'].obesity)
```

### 2-4. 카이 제곱 검정을 이용해 위 저체중/정상/비만 체중 빈도가 남성 vs. 여성 간에 차이 나는지 검정하시오. (p값 계산)

```python
stats.chi2_contingency(pd.crosstab(data[data.age_group2 == '청년'].sex, data[data.age_group2 == '청년'].obesity))
```
