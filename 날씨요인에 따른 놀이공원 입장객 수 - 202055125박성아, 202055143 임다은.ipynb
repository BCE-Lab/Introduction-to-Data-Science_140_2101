```python
dust1 = pd.read_csv('dust.csv', encoding = 'CP949')
dust2 = pd.read_csv('dust1.csv', encoding = 'CP949')
dust3 = pd.read_csv('dust2.csv', encoding = 'CP949')
dust = pd.concat([dust3,dust2,dust1])
dust = dust.set_index('일시')
```

```python
climate1 = pd.read_csv('climate.csv', encoding = 'cp949',index_col = '일시' )
climate2 = pd.read_csv('climate1.csv', encoding = 'cp949',index_col = '일시' )
climate3 = pd.read_csv('climate2.csv', encoding = 'cp949',index_col = '일시' )
climate = pd.concat([climate3, climate2,climate1])
climate
```

```python
climate = climate.drop(['지점','지점명','최저기온 시각(hhmi)','최고기온 시각(hhmi)','최대 순간풍속 시각(hhmi)','최대 순간 풍속 풍향(deg)'],axis = 1)
dust = dust.drop(['지점','지점명'], axis = 1)
weather = pd.merge(dust, climate,how = 'outer', left_index = True, right_index = True)


```

```python
w = weather[weather.dust>=300].index[0]
```

```python
weather.drop(w,inplace = True)
```

```python
def cleaning(x) :
    if str(x).find(","):
        return str(x).replace(",","")
    else:
        return str(x)
```

```python
data_final = data[['date','day','people_T','Weekend','month']]
data_final = data_final.set_index('date')

d_ff = pd.merge(data_final, weather, left_index = True, right_index = True)
d_ff['people_T']=d_ff['people_T'].apply(cleaning).apply(lambda x: float(x))
d_ff.dropna(inplace = True)
d_ff
```

```python
def floating(x) :
    if str(x).find("."):
        return str(x).split(".")[0].replace(".","")
    else:
        return str(x)
```

```python
d_ff['temperature_f']= d_ff['temperature_m'].apply(floating).apply(lambda x: int(x))
d_ff
```

```python
def remove_outlier(d_cp, column):
    column_data = d_cp[column]
    q_25 = np.percentile(column_data.values, 25)
    q_75 = np.percentile(column_data.values, 75)
    iqr = q_75-q_25
    iqr = iqr *1.5
    lowest = q_25-iqr
    highest = q_75+iqr
    outlier_index = column_data[(column_data<lowest) | (column_data>highest)].index
    print(len(outlier_index))
    d_cp.drop(outlier_index, axis = 0, inplace = True)
    print(d_cp.shape)
    return d_cp
```

```python
d_ff_d = remove_outlier(d_ff, 'people_T')
```
