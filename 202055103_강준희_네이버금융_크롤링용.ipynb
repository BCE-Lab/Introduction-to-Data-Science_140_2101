```python
url = 'https://finance.naver.com/sise/sise_index_day.nhn?code=KOSPI'
headers = {'user-agent':'Mozilla/5.0 (Macinthosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
req = requests.get(url, headers = headers)

soup = BeautifulSoup(req.text, 'html.parser')
last_page = int(soup.select_one('td.pgRR').a['href'].split('=')[-1])

df = None
for page in range(1, last_page):
    req = requests.get(f'{url}&page={page}', headers = headers)
    df = pd.concat([df, pd.read_html(req.text, encoding = 'euc-kr')[0]], ignore_index = True)

df.dropna(inplace=True)
df.reset_index(drop = True, inplace=True)
df
# last_page
```

```python
df.to_csv('KOSPI.csv', index = None, na_rep = 'NaN', encoding='euc-kr')
```
